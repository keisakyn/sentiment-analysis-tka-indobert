```python
import pandas as pd
```

```python
df = pd.read_csv(f"dataset_tka (1).csv")
```

```python
df.info()
```

```python
df = df.drop_duplicates(subset=['full_text'])
df = df.dropna(subset=['full_text'])
df.reset_index(drop=True, inplace=True)
print("Jumlah data setelah hapus duplikat:", len(df))
display(df)
```

```python
# Membuat dataframe sementara berisi data yang TIDAK mengandung 'TKA'
df_dropped = df[~df['full_text'].str.contains('TKA' , case=False, na=False)]

# Menampilkan data yang akan dibuang
print(f"Jumlah data yang akan di-drop: {len(df_dropped)} baris")
display(df_dropped[['username', 'full_text']].head(70)) # Melihat 20 sampel teratas
```

```python
import re
promo = [
    "harga", "tryout", "bimbel", "les", "kelas gratis",
    "daftar sekarang", "diskon", "promo", "open kelas",
    "pendaftaran", "link di bio", "hubungi", "wa",
    "wts","wtb","garansi","pricelist",
    "whatsapp", "t.me", "telegram", "zoom class",
    "ruangguru", "zenius", "brain academy",
    "kode promo", "tutor", "privat", "booking",
    "join kelas", "kelas intensif", "ebook gratis",
    "jualan", "open jasa", "admin", "slot terbatas",
    "detik", "kompas", "cnn", "tribun", "tempo",
    "liputan6", "kumparan", "tvone", "metrotv", "china",
    "pemerintah","pemimpin","dzalim","akbar","prabowo",
    "asing","kritik", "jokowi", "tirto","pejabat","antek",
    "asing","stafsus", "korupsi", "kpk","kemnaker","disnaker",
    "disnakertrans","menaker","menakertrans","imigrasi","visa",
    "pemerasan","tersangka","buruh","tki","tkw","deportasi","outsourcing"
]

def cek_spam(text):
    text = str(text).lower()
    for word in promo:
        if re.search(r'\b' + re.escape(word) + r'\b', text):
            return True
    return False
```

```python
df_clean = df[~df['full_text'].apply(cek_spam)]
df_clean.reset_index(drop=True, inplace=True)

print("Jumlah data awal:", len(df))
print("Jumlah data setelah hapus promosi:", len(df_clean))
```

```python
#hapus link
df_clean = df_clean[~df_clean['full_text'].str.contains("http", case=False)]
#hapus twit pendek
df_clean = df_clean[df_clean['full_text'].str.len() > 10]
```

```python
#reset index
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
# Konversi kolom "created_at" ke tipe datetime
df_clean["created_at"] = pd.to_datetime(df["created_at"])
# Mengurutkan data dari yang paling lama ke yang terbaru
df_sorted = df_clean.sort_values(by="created_at", ascending=True)

print(df_sorted[["created_at"]].head())  # 5 data teratas
print(df_sorted[["created_at", "full_text"]].tail(70))  # 5 data terbawah
```
